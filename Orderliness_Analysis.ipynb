```python
import numpy as np
import matplotlib.pyplot as plt
numOfDays=10  # Will simulate data for these many number of days
cycle=48
t=np.arange(cycle*numOfDays) # the time in the time series
```

```python
# Having a simulated dataset of regular meals.
# 'regular' is an array of 'number of days'*'48' 0s to begin with. Each time a person has a meal at any xth time...
#... we increment the value of regular[x] by 1
# Now imagine 'regular' as a time series where value at index i represents the number of meals had at...
#... i*0.5 hrs from midnight of the starting date.

everyday=np.array([18,26,40])
regular=np.zeros(t.shape[0])
for i in range(0,numOfDays):
    for j in everyday:
        regular[int(j)-1]+=1
    everyday+=cycle
print(regular)
```

```python
plt.figure(num=1,figsize=(15,5))
plt.bar(np.arange(t.shape[0]),regular)
plt.title('Time series regular for 100 days)')
plt.xlabel('Time (absolute)')
plt.ylabel('Amplitude (how many meals have been had at this hour)')
plt.show()
```

```python
# Similarly defining irregular meals
irregular=np.zeros(t.shape[0])
for i in range(0,numOfDays):
    randomMeal=(i*48)+np.floor(np.random.rand(3)*48)   # rounding off meals to the nearest half integer.
                                                # using 'floor' because 'around' can give index 48 (out of bounds)
    for j in randomMeal:
        irregular[int(j)]+=1
print(irregular)
```

```python
plt.figure(num=1,figsize=(15,5))
plt.bar(np.arange(t.shape[0]),irregular)
plt.title('Time series irregular for 100 days (bar graph)')
plt.xlabel('Time (absolute)')
plt.ylabel('Amplitude (how many meals have been had at this hour)')
plt.show()
```

```python
# Fourier transforming to frequency domain
fft_regular=np.fft.fft(regular)
fft_irregular=np.fft.fft(irregular)

# When you do a fourier transform(FT), the time series is converted to the frequency domain.
# The value stored in 'fft_regular'/'fft_irregular' is actually a complex value as seen below
print(fft_irregular[5])

# Think of the wave example again. When we do a FT, we are doing frequency decomposition of the given...
# time series signal. If the time series is regular, it will be made of a smaller number of waves...
#... making big contributions. If the time series is irregular, FT will have a large number of waves making...
# approximately uniformly distributed contribution.
# Because this is a wave, it has an amplitude and a phase. That is why we have a complex number.
# The amplitude of a compelx number a+ib is (a^2+b^2)^1/2
```

```python
plt.figure(num=3,figsize=(15,5))
plt.plot(np.arange(t.shape[0]),np.absolute(fft_regular))
plt.title('Frequency series regular')
plt.xlabel('Frequency')
plt.ylabel('Amplitude (How much a freq contributes to the regularity)')
plt.show()
plt.figure(num=4,figsize=(15,5))
plt.plot(np.arange(t.shape[0]),np.absolute(fft_irregular))
plt.title('Frequency series irregular')
plt.xlabel('Frequency')
plt.ylabel('Amplitude (How much a freq contributes to the regularity)')
plt.show()

# You can see how the regular guy has a much organised FT than the irregular guy.
# You can find orderiless by doing a simple STD exercise on these FT values
```

```python
# The actual constituent frequencies can be found as follows:
freq=np.fft.fftfreq(t.shape[0])
# your the values of 'fft_regular'/irrgular correspond index wise to the frequencies in the 'freq' array
print(freq[:5],freq[-5:])
```

```python
# You can observe that some of the frequencies are negative. This is just a mathematical convention devised to make...
#... fourier transforms easier. For your analysis, you can just ignore the FT values associated with negative frequencies
fft_regular=fft_regular[freq>0.0]   # keeping values corresponding to positive frequencies only
fft_irregular=fft_irregular[freq>0.0]
freq=freq[freq>0.0]

plt.figure(num=5,figsize=(15,5))
plt.plot(freq*cycle,np.absolute(fft_regular),color='b',label='reg')
plt.plot(freq*cycle,np.absolute(fft_irregular),color='r',label='irreg')
plt.xlabel('Frequencies')
plt.ylabel('Amplitude of contribution')
plt.legend()
plt.show()
```

```python
# Defining a very simple orderliness measure. You can ideally define many different measures
# My definition is as follows-
# Find the average amplitude. Count the number of spikes above the average amplitude and compare
# Another analysis could be to find the maximum amplitude and to find the number of spikes above some fraction...
#... f (0<f<1) of the max amplitude.
# If we have more spikes, it is less orderly. If we have less spikes it is more orderly.
# Hence, orderliness score = 1/(number of spikes selected from above methods)

# amp is the amplitude array for each frequency from the FFT
def orderliness_1(amp):
    avg=np.mean(amp)
    spikes=0
    for i in range(1,amp.size-1):
        # finding spikes by checking of current value is bigger than neighboring
        if amp[i]>amp[i-1] and amp[i]>amp[i+1]:
            spikes+=1
    return 1.0/spikes

# amp  is the same as above, f is the fraction as mentioned above
def orderliness_2(f,amp):
    max_amp=np.amax(amp)
    spikes=0
    for i in range(1,amp.size-1):
        # finding spike and checking if above fraction
        if amp[i]>amp[i-1] and amp[i]>amp[i+1]  and amp[i]>f*max_amp:
            spikes+=1
    return 1.0/spikes


print('Method 1 orderliness score')
print('Regular: '+str(orderliness_1(fft_regular))+'  |  Irregular: '+str(orderliness_1(fft_irregular)))
print('Ratio of score: '+str(orderliness_1(fft_regular)/orderliness_1(fft_irregular)))
print('Method 2 orderliness score, Fraction f=0.5')
print('Regular: '+str(orderliness_2(0.5,fft_regular))+'  |  Irregular: '+str(orderliness_2(0.5,fft_irregular)))
print('Ratio of score: '+str(orderliness_2(0.5,fft_regular)/orderliness_2(0.5,fft_irregular)))
print('Method 2 orderliness score, Fraction f=0.25')
print('Regular: '+str(orderliness_2(0.25,fft_regular))+'  |  Irregular: '+str(orderliness_2(0.25,fft_irregular)))
print('Ratio of score: '+str(orderliness_2(0.25,fft_regular)/orderliness_2(0.25,fft_irregular)))
print('Method 2 orderliness score, Fraction f=0.75')
print('Regular: '+str(orderliness_2(0.75,fft_regular))+'  |  Irregular: '+str(orderliness_2(0.75,fft_irregular)))
print('Ratio of score: '+str(orderliness_2(0.75,fft_regular)/orderliness_2(0.75,fft_irregular)))
```

## You can clearly see that method 1 gives a very bad estimate and Method 2 with f=0.3 gives a very good estimate.
