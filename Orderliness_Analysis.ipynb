```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Having a simulated dataset of regular meals.
everyday=np.array([18.0,26.0,40.0])
regular=np.array([])
regular=everyday
for i in range(0,100):
#    everyday=everyday+48.0
    regular=np.concatenate((regular,everyday))
    
amplitude=np.ones(regular.size)
print(regular[:9])
```

```python
plt.figure(num=1,figsize=(15,5))
plt.scatter(regular,amplitude,s=10)
plt.title('Time series regular')
plt.xlabel('Time (absolute)')
plt.ylabel('Number of meals (can be changed to amount of food)')
plt.show()
```

```python
# Having irregular meal times simulation
irregular=np.array([])
irregular=np.sort(np.random.rand(3)*48.0)   # because random numbers returns in the range [0,1)
for i in range(0,100):
    irregular=np.concatenate((irregular,np.sort(np.random.rand(3)*48)))
print(irregular[:9])
```

```python
plt.figure(num=2,figsize=(15,5))
plt.scatter(irregular,amplitude,s=5)
plt.title('Time series irregular')
plt.xlabel('Time (absolute)')
plt.ylabel('Number of meals (can be changed to amount of food)')
plt.show()
```

```python
# Fourier transforming to frequency domain
fft_regular=np.real(np.fft.rfft(regular))
fft_irregular=np.real(np.fft.rfft(irregular))
```

```python
plt.figure(num=3,figsize=(15,5))
plt.scatter(fft_regular,np.ones(fft_regular.size),s=15)
plt.title('Frequency series regular')
plt.xlabel('Frequency')
plt.ylabel('Number of meals (can be changed to amount of food)')
plt.show()
```

```python
plt.figure(num=4,figsize=(15,5))
plt.scatter(fft_irregular,np.ones(fft_irregular.size),s=1)
plt.title('Frequency series irregular')
plt.xlabel('Frequency')
plt.ylabel('Number of meals (can be changed to amount of food)')
plt.show()
```
